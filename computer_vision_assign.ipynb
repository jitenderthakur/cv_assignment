## 3. How to change find the shape of the Image?

```python
import cv2
 
img = cv2.imread('C:/Users/jitender thakur/Downloads/a.png', cv2.IMREAD_UNCHANGED)
 
print('Original Dimensions : ', img.shape)
 
scale_percent = 60 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 9. Read any image and apply the Blurring technologies which available in opencv then show the results in matplotlib library

```python
import cv2
import matplotlib.pyplot as plt

image= cv2.imread('C:/Users/jitender thakur/Downloads/a.png')

gausBlur = cv2.GaussianBlur(image, (5,5),0) 

plt.imshow(gausBlur)
plt.title('gausBlur')
plt.show()

```

```python
image_blurred = cv2.blur(src=image, ksize=(50, 50))
plt.imshow(image_blurred)
plt.title('Blur')
plt.show()
```

## 10. Read the saved video file then create output folder and store the frames in the local output folder.

```python
import cv2
vidcap = cv2.VideoCapture('trucks.mp4')
success,image = vidcap.read()
count = 0
success = True
while success:
    success,image = vidcap.read()
    cv2.imwrite("frame%d.jpg" % count, image)    
    if cv2.waitKey(10) == 27:                    
        break
    count += 1
```

## 11. What is dilation and erosion and give some example then apply the techlogies to some image?



```python
"Dilation adds pixels to the boundaries of objects in an image, while erosion removes pixels on object boundaries"
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
# Reading the input image
img = cv2.imread('C:/Users/jitender thakur/Downloads/a.png', 0)
 
kernel = np.ones((5,5), np.uint8)

img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)
 

plt.imshow(img_erosion)
plt.title('Erosion')
plt.show()
plt.imshow(img_dilation)
plt.title('Dilation')
plt.show()
 
```

## 14. How can you do image compression in opencv ?

```python
# import openCV library for image handling
import cv2

# read image to be resized by imread() function of openCV library
img = cv2.imread('C:/Users/jitender thakur/Downloads/a.png')
print(img.shape)

# set the ratio of resized image
k = 5
width = int((img.shape[1])/k)
height = int((img.shape[0])/k)

scaled = cv2.resize(img, (width, height), interpolation=cv2.INTER_AREA)
print(scaled.shape)

plt.imshow(scaled)
```
